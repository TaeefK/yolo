```python
import cv2
import os

# Function to extract frames from videos in a folder
def extract_frames(video_folder, output_folder):
    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # List video files in the folder
    video_files = [f for f in os.listdir(video_folder) if f.endswith(('.mp4', '.avi'))]  # Adjust file types as needed

    # Loop through each video file
    for video_file in video_files:
        video_path = os.path.join(video_folder, video_file)
        cap = cv2.VideoCapture(video_path)
        frame_count = 0

        while True:
            success, frame = cap.read()
            if not success:
                break

            # Save the frame as an image
            cv2.imwrite(f"{output_folder}/frame_{video_file}_{frame_count:04d}.jpg", frame)
            frame_count += 1

        cap.release()
        print(f"Extracted {frame_count} frames from {video_file}")

# Folder paths for autistic and non-autistic videos
autistic_folder = '/content/autistic_videos/autistic'
nonautistic_folder = '/content/nonautistic_videos/nonautistic'

# Output folders for extracted frames
autistic_output_folder = '/content/dataset/images/train/autistic'
nonautistic_output_folder = '/content/dataset/images/train/nonautistic'

# Extract frames from both autistic and non-autistic videos
extract_frames(autistic_folder, autistic_output_folder)
extract_frames(nonautistic_folder, nonautistic_output_folder)
```
